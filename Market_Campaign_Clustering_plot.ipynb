```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
principal_components = pd.DataFrame(pca.fit_transform(df_input), columns=['PC1', 'PC2'])
principal_components
```

```python
plt.scatter(principal_components['PC1'], principal_components['PC2'])
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title(f"Data Scatter After PCA")
plt.show()
```

```python
from sklearn.cluster import KMeans

distortions = []
for i in range(2, 10):
  kmeans_model = KMeans(n_clusters = i)
  cluster_label = kmeans_model.fit_predict(principal_components)
  distortions.append(kmeans_model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(range(2, 10), distortions, marker='o')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeanModel = KMeans(n_clusters=2, init='k-means++', random_state=0)
clusters = kmeanModel.fit_predict(principal_components)
```

```python
dataset['k_means']=kmeanModel.predict(principal_components)
```
